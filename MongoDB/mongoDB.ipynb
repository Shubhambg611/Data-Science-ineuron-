```python
#to fetch only one record
mycol.find_one()
```

```python
#to fetch all record
for x in mycol.find():
    print(x)
```

```python
#find specific record
mycol.find_one({"name":"shubham"})
```

```python
# to find multiple record 
for x in mycol.find({"Add":"Mumbai"}):
    print(x)
```

```python
# Sorting operation
for x in mycol.find().sort("name"):
    print(x)
```

```python
mycol.delete_one({"name":"chirag"})
```

```python
for x in mycol.find().limit(3):
    print(x)
```

```python
myquery={"name":"shubham"}
newvalue={"$set":{"name":"Shubham"}}
```

```python
mycol.update_one(myquery,newvalue)
```

```python
mycol.update_one({"last name":"Gadhave"},{"$set":{"last name":"GADHAVE"}})
```

```python
# comparison operator
for x in mycol.find({"salary":{"$gt":10000000}}):
    print(x)
```
