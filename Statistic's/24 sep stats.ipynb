```python
import matplotlib.pyplot as plt 
import scipy.stats as stats
import pylab

def plot_data(sample):
    plt.figure(figsize=(10,6)) # deciding the size lenght and width
    plt.subplot(1,2,1)
    sns.histplot(sample,kde=True)
    plt.subplot(1,2,2)
    stats.probplot(sample,dist="norm",plot=pylab)
    plt.show()
```

```python
plot_data(log_normaldata2)
```

```python
plot_data(np.log(log_normaldata2))
```

### Central limit theoram

The Central Limit Theorem (CLT) is a statistical theory that states that if a sample size is large enough, the mean of the samples will be close to the mean of the population.

* Suppose you have a population with any shape of probability distribution (not necessarily normal), and you draw random samples of a fixed size from this population. As the sample size increases, the distribution of the sample means approaches a normal distribution, regardless of the shape of the original population distribution.
