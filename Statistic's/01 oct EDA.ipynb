<b><u>EDA</u></b>

EDA is a fundamental process in many data science and analysis tasks. It's a first step in any data 
analysis.<br><br>
EDA is a philosophy that allows data analysts to approach a database without assumptions. It identifies general patterns in the data, including outliers and unexpected features.<br><br>
<b>EDA is a process that involves:</b><br> 
* Data collection<br>
* Cleaning the dataset<br>
* Identifying correlated variables<br>
* Choosing the right statistical methods<br>
