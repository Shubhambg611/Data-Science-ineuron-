## Z test

What Is a Z-Test?
A z-test is a statistical test used to determine whether two population means are different when the variances are known and the sample size is large.

The test statistic is assumed to have a normal distribution, and nuisance parameters such as standard deviation should be known in order for an accurate z-test to be performed.



```python
sample_mean = 110
popoulation_mean = 100
population_std = 15
sample_size = 50
alpha = 0.05
```

```python
confidence_interval = 1-0.05
```

```python
z_score = (sample_mean-popoulation_mean)/ (population_std/np.sqrt(50))
print(z_score)
```

```python
z_critical = stats.norm.ppf(confidence_interval) # z critical value
```

```python
if z_score>z_critical:
    print("null hypothesis is rejected")
else:
    print("Fail to reject the null hypothesis")
```

Calculating the P-value:<br>
the probability under the assumption of no effect or no difference (null hypothesis), of obtaining a result equal to or more extreme than what was actually observed.

## T-test

A T-test is a statistical test that is used to compare the means of two groups. It is often used in hypothesis testing to determine whether a process or treatment actually has an effect on the population of interest, or whether two groups are different from one another.<br>

<b>when to apply</b>
* limited sample size (n<30)<br>
* if population statndard deviation is missing then T test is performed

<b>Types:-</b> <br>
one-sample t test:- (with respect to one independent sample)<br>
two- sample t test:- (with respect to two independent sample)<br>
paried t-test:- (two sample from the same populatoin on a different time interval)


#### <i><b>if the p<0.05 which mean the probablity is below 95% then null hypothesis is rejected</b><i>  <br>

: The weights of 25 obese people were taken before enrolling them into the nutrition camp. The population mean weight is found to be 45 kg before starting the camp. #### <b>Solution</b>
n = 25<br>
population mean () = 45kg <br>
sample mean (x) = 75<br>
sample S.D = 25<br>

Null hypo :-    =45kg <br>
alternate hypo:-        45kg <br>


it will be two tail test becouse they have not given any info like it should be less or greater. 

```python
n = 25
population_mean = 45
sample_mean = 75 
sample_std = 25
```

```python
t_stats = (sample_mean-population_mean)/(sample_std/np.sqrt(n))
print(t_stats)
```

```python
df = n-1
print(df)
CI=0.95
alpha=0.05
```

```python
critiacal_value = stats.t.ppf(CI,df)
critiacal_value
```

```python
print("t-stats",t_stats)
print("critical value",critiacal_value)
```

```python
if t_stats>critiacal_value:
    print("reject the null hypo")
else:
    print("accept the null hypo")
```

now same thing with p value

```python
p_value = 1 - stats.t.cdf(t_stats,df)
p_value
```

```python
if p_value<=0.05:
    print("reject the null hypo")
else:
    print("fail to reject the null hypo(Accept)")
```

### <b>Chi Squre Test</b><br>

Chi squared testes Perform on categorical data.<br>

<b>A chi-square test is a statistical test used to compare observed results with expected results</b><br>

The chi-square test is used to examine the association between two categorical variables. The null hypothesis states that there is no relationship between the two variables, while the research hypothesis states that there is a relationship.

```python
 # crosstab use to Segregate the categories.
 import pandas as pd
 dataset_table = pd.crosstab(dataset["sex"],dataset["smoker"])
```

```python
dataset_table #observed value
```

```python
dataset_array = dataset_table.values # convert above table into array
observed_data = dataset_array
```

```python
import scipy.stats as stats
```

```python
#expected value
stats_test,p,dof,expected_value=stats.chi2_contingency(dataset_array)
```

```python
stats_test
```

<b>DOF</b> =  (number of rows - 1) * (number of column-1)

```python
no_of_column = dataset_table.iloc[0:2,0]
```

```python
len(no_of_column)
```

```python
no_of_rows = dataset_table.iloc[0,0:2]
```

```python
len(no_of_rows)
```

```python
dof=((len(no_of_column)-1)*(len(no_of_rows)-1))
```

```python
dof
```

```python
for o,e in zip(observed_data,expected_value):
    print(o,e)
    
```

```python
from scipy.stats import chi2
chisqure = sum([(o-e)**2/e for o,e in zip(observed_data,expected_value)]) 
```

```python
chisqure
```

```python
chisqurestats = chisqure[0]+chisqure[1]
```

```python
chisqurestats
```

```python
alpha = 0.05
ci = 0.95
```

```python
critical_chi_value = chi2.ppf(ci,df=dof)
```

```python
critical_chi_value
```

```python
if chisqurestats>=critical_chi_value:
    print(" reject the null hyphothesis, there is a relatiuonship between two categorical variable")
else:
    print("accept the null hyphothesis there is no relation between two categorical variable")
```
