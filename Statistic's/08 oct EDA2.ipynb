# Outlier detection and handling 
    
An outlier is a data point that is far from the average value of a group of statistics. Outliers can be caused by measurement variability, experimental error, or novel data. They can also be an indication of exciting possibilities, but can also cause problems in statistical analyses. <br>

## <u><b>Trimming of data :-</b></u><br>
may have normally distributed or non normal distributed data if there are 1/2 outlier we simply remove then from the data 
<BR><BR>
#### <b>For Normally distributed data</b>
<b>to trim use 3 standard deviation data and remainiing trim</b><br><br>
<b>trim</b> = mean +- 3rd standard deviation     (removing the tails of the normal distribution)

 

 #### <b>For Non Normally distributed data</b><br>
 
<BR>
 <br>Use 5 number summury for handling the non noraml distribution outliers<br>
    * plot box plot and lower than lower fence and higher than upper fence trim those data points
    * % technique for triming tha data 

## <u><b>Capping of the data</b></u><br>
Capping. In this technique, wecap our outliers data and make the limit i.e, above a particular value or less than that value, all the values will be considered as outliers, and the number of outliers in the dataset gives that capping number.<br>

in capping the outliers converted into upper fence value or lower fence value<br>

# capping

```python
upper_limit = df['cgpa'].mean() + 3* df['cgpa'].std()
lower_limit = df['cgpa'].mean() - 3* df['cgpa'].std()
```

```python
df['cgpa'] = np.where(df["cgpa"]>upper_limit,upper_limit,df['cgpa'])
```

```python
df['cgpa'] = np.where(df["cgpa"]<lower_limit,lower_limit,df['cgpa'])
```

## IQR

```python
sns.boxplot(df['placement_exam_marks'])
```

```python
percentile25 = df['placement_exam_marks'].quantile(0.25)
percentile75 = df['placement_exam_marks'].quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
iqr= percentile75-percentile25
iqr
```

```python
lower_fence = percentile25 - 1.5 * iqr
upper_fence = percentile75 + 1.5 * iqr
```

```python
lower_fence
```

```python
upper_fence
```

```python
 df[df['placement_exam_marks']>upper_fence]
```

```python
df[df['placement_exam_marks']<lower_fence]
```

```python
new_df = df[df['placement_exam_marks'] <= upper_fence]
```

```python
sns.boxplot(new_df['placement_exam_marks'])
```
