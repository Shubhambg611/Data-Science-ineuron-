```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np 
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
x
```

```python
y
```

```python
#plot x and y axis
```

```python
plt.scatter(x,y,c='r') # c is used for colour of plot 
plt.xlabel(" X axis")
plt.ylabel("y axis")
plt.title("Graph In 2D")
plt.savefig("2d.png")
```

```python
y=x*x
```

```python
#line plot
plt.plot(x,y,c="r")
```

```python
plt.plot(x,y,"r*",linestyle="dashed",linewidth=2)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("2D Diagonal")
```

Creating plot subplot

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(y,x,'g--')
plt.subplot(2,2,3)
plt.plot(y,x,'b--')
plt.subplot(2,2,4)
plt.plot(x,y,'b*')
```

## box plot

```python
data = [np.random.normal(0,std,100) for std in range(1,4)] #list comprasion 
```

```python
plt.boxplot(data)
```

## pie chart

```python
#data for plot
labels = 'python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','red','lightskyblue']
explode = (0.1,0,0,0)

#plot

plt.pie(sizes,explode=explode,labels=labels,colors=colors, autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
